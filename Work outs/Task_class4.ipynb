# Preprocessing

Handling Missing Values

```python
df.isnull().sum()
```

```python
#Imputation
df['Price ($)'].fillna(df['Price ($)'].mean(),inplace = True)
```

```python
#removing null values
df.dropna(inplace = True)
```

# Visualization

1. Count plot

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Count")
plt.show()
```

```python
sns.countplot(x='Dealer_Region', data=df)
plt.title("Region wise Count")
plt.show()
```

```python
sns.countplot(x='Engine', data=df)
plt.title("Engine wise Count")
plt.show()
```

```python
sns.countplot(x='Transmission', data=df)
plt.title("Transmission wise Count")
plt.show()
```

2. Correlation heatmap

```python
df.corr(numeric_only=True)
```

```python
df.dropna(subset=['Phone'], inplace=True)
```

```python
selected_columns = df[['Annual Income', 'Price ($)']]
correlation_matrix_selected = selected_columns.corr()
sns.heatmap(correlation_matrix_selected, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap: Annual Income vs Price ($")
plt.show()
```

3. Bar Plot

```python
sns.barplot(x='Price ($)', y='Company', data=df)
plt.title("Company wise Price of Car")
plt.show()
```

4. Pie chart

```python
color_counts = df['Color'].value_counts()
plt.pie(color_counts, labels=color_counts.index, autopct='%1.1f%%')
plt.title("Pie Chart of Car Colors")
plt.show()
```

5. Scatter plot

```python
sns.scatterplot(x="Annual Income", y="Price ($)", data=df)
plt.title("Annual Income vs Price of Car")
plt.show()
```
