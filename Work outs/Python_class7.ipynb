# 1. Find the Factorial of a Number


```python
def find_factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
print(find_factorial(5))
```

# 2. Find the Largest and Smallest Number in a List


```python
def find_min_max(numbers):
    if not numbers:
        return None

    smallest = numbers[0]
    largest = numbers[0]

    for num in numbers:
        if num < smallest:
            smallest = num
        if num > largest:
            largest = num

    return smallest, largest
```

```python
print(find_min_max([85,44,32,11,7,910]))
```

# 3. Fibonacci Sequence up to N terms


```python
def fibonacci_sequence(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
        a = b
        b = a+b
```

```python
print(fibonacci_sequence(10))
```

```python
print(fibonacci_sequence(4))
```

# 4. Create a Square Frame Pattern


```python
def square_frame(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()
```

```python
print(square_frame(4))
```

# 5. Sum of Digits in a Number


```python
def sum_of_digits(n):
    total = 0
    for digit in str(n):
        total += int(digit)
    return total
```

```python
print(sum_of_digits(453))
```
