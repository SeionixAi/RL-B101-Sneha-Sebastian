6. break, continue, and pass
Write a program that:
* Demonstrates break, continue, and pass

```python
for i in range(1, 11):
    if i == 3:
        continue
    if i == 5:
        break
    if i == 7:
        pass
    print(i)
```

7. Recursion
Write a recursive function that:
* Finds the factorial of a number

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```
