# 1. Write a Python program to count the number of consonants in a given string.



```python
def count_consonants(text):
    vowels = "aeiou"
    count = 0
    for char in text.lower():
        if char.isalpha() and char not in vowels:
            count += 1
    return count
```

```python
print(count_consonants('Artificial Intelligence'))
```

# 3. Write a Python program to remove all duplicate characters from a string.



```python
def remove_duplicates(text):
    result = ""
    for char in text:
        # Step 1: If the character is not already in our result, add it
        if char not in result:
            result += char
    return result
```

```python
print(remove_duplicates('Computer Programming'))
```

# 4. Write a Python program to count the frequency of each character in a string.



```python
def count_char_frequency(text):
    frequency = {}
    for char in text:
        # Step 1: If char is in dictionary, increase count
        if char in frequency:
            frequency[char] += 1
        # Step 2: If char is not in dictionary, start it at 1
        else:
            frequency[char] = 1
    return frequency
```

```python
print(count_char_frequency('Machine learning'))
```

# 2. Write a Python program to find the first non-repeating character in a string.



```python
def first_non_repeating(text):
    # Step 1: Convert the text to lowercase
    lower_text = text.lower()

    # Step 2: Count the frequency of each character
    char_count = {}
    for char in lower_text:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    # Step 3: Find the first character with count 1
    for char in lower_text:
        if char_count[char] == 1:
            return char

    return None
```

```python
print(first_non_repeating('programming'))
```

```python
print(first_non_repeating('nano'))
```
