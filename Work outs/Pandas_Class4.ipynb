# 5. Handling missing values

```python
df.isnull().sum() # Count nulls
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True) # Fill with mean
```

```python
df.dropna(inplace=True) # Drop rows with any nulls
```

# 6. Removing Duplicates


```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

# 7. Selecting columns

```python
df['Name'] # Single column (Series)
```

```python
df[['Name', 'Age']] # Multiple columns (DataFrame)
```

# 8. Filtering rows

```python
df[df['Age'] > 30] # Age > 30
```

```python
df[df['Sex'] == 'female'] # All females
```

```python
df[(df['Age'] > 30) & (df['Sex'] == 'male')] # Compound filter
```

# 9. Sorting data

```python
df.sort_values(by='Age', ascending=True)
```

```python
df.sort_values(by=['Pclass', 'Fare'], ascending=[True, False])
```

# 10. Creating new column

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

# 11. Dropping columns/rows

```python
df.drop('Ticket', axis=1, inplace=True) # Drop column
```

```python
# df.drop([0, 1], axis=0, inplace=True) # Drop rows by index
# (The error KeyError: '[0, 1] not found in axis' means that the DataFrame no longer has rows with index labels 0 and 1.
#This happened because previous operations like df.dropna() and df.drop_duplicates() modified the DataFrame and removed those original index labels.
#If you want to drop the first two rows of the current DataFrame, you need to reference their current index labels.
#I've updated the code to drop the first two rows based on their current labels.)
df.drop(df.index[:2], axis=0, inplace=True) # Drop the first two current rows
```

# 12. Remaining columns

```python
df.rename(columns={'SibSp': 'SiblingsSpouses'}, inplace=True)
```

# 14. Grouping and aggregation

```python
df.groupby('Pclass')['Fare'].mean()
```

# 15. Aggregating Multiple Columns

```python
df.groupby('Pclass').agg({
'Age': ['mean', 'min'],
'Fare': ['max', 'mean']
})
```

# 16. Apply Custom Functions

```python
df['AgeSquared'] = df['Age'].apply(lambda x: x**2)
```

# 17. String operations

```python
df['Name'].str.lower()
```

# 18. Mapping and replacing values

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df['Embarked'].replace({'S': 'Southampton', 'C': 'Cherbourg'}, inplace=True)
```

# 19. Combining Columns (concatenation)

```python
df['FullName'] = df['Name'] + ' (' + df['Sex'].astype(str) + ')'
```

# 20. Merging Data frames

```python
df2 = pd.read_csv('/content/tested.csv')
```

```python
df = pd.merge(df, df2, on='PassengerId', how='left')
```

# 21. Concatenating dataframes

```python
df_combined = pd.concat([df1, df2], axis=0) # Stack vertically
```

# 22. Pivot table

```python
pd.pivot_table(df, values='Fare_x', index='Pclass_x', columns='Sex_x', aggfunc='mean')
```

# 23. Resetting and setting index

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.set_index('PassengerId', inplace=True)
```

# 24. Conditional column creation

```python
df['AgeGroup'] = df['Age_x'].apply(lambda x: 'child' if x < 16 else 'adult')
```
