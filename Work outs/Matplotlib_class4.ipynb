```python
import matplotlib.pyplot as plt
import numpy as np
```

# Sample data

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

# 1. Line plot

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

# 2.Scatter plot

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

# 3.Bar chart

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

# 4.Horizontal bar chart

```python
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.show()
```

# 5. Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram")
plt.show()
```

# 6.Box plot

```python
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

# 7. Pie chart

```python
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

# 8. Multiple lines on one plot

```python
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

# 9.Sub plot

```python
fig, axs = plt.subplots(2, 1)
axs[0].plot(x, np.sin(x))
axs[1].plot(x, np.cos(x))
plt.suptitle("Subplots Example")
plt.show()
```

# 10.Changing colour and style

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title("Styled Line")
plt.show()
```

# 11. Adding labels and titles

```python
plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Labeled Plot")
plt.show()
```

# 12 Fill between

```python
plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.title("Fill Between")
plt.show()
```

# 13.Annotate points

```python
plt.plot(x, y)
plt.annotate("Peak", xy=(np.pi/2, 1), xytext=(2, 1.2),
arrowprops=dict(facecolor='black', shrink=0.05))
plt.title("Annotation Example")
plt.show()
```

# 14. Save figure

```python
plt.plot(x, y)
plt.savefig("line_plot.png")
```

# 15.Gridlines

```python
plt.plot(x, y)
plt.grid(True)
plt.title("Gridlines Example")
plt.show()
```
