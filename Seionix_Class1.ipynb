# PART 2: NumPy Coding Tasks

```python
import numpy as np
```

1.Create Arrays

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

2.Array Operations

```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])
```

```python
print(a + b)
print(a * b)
```

3.Zeros and Ones

```python
print(np.zeros((2, 3)))
print(np.ones((3, 3)))
```

```python
data = np.array([10, 20, 30, 40])
```

```python
print("Mean:", np.mean(data))
print("Max:", np.max(data))
print("Min:", np.min(data))
```
