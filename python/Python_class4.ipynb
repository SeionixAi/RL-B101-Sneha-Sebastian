# 1. Write a Python program to count the number of vowels in a given string.

```python
def count_vowels(text):
  vowels = "aeiou"
  count = 0
  for char in text.lower():
    if char in vowels:
      count += 1
  return count
```

```python
# Example
print(count_vowels("Artificial Intelligence"))
```

# 2. Write a Python program to check whether a given string is a palindrome.

```python
def is_palindrome(text):
   # Step 1: Convert the input text to lowercase
   lower_text = text.lower()

   # Step 2: Reverse the lowercase text
   reversed_text = lower_text[::-1]

   # Step 3: Compare the original and reversed text
   if lower_text == reversed_text:
      return True
   else:
      return False
```

```python
# Example usage
result1 = is_palindrome("madam")
print(result1)   # True
```

```python
result2 = is_palindrome("python")
print(result2)   # False
```

# 3. Write a Python program to check whether two strings are anagrams.

```python
def is_anagram(str1, str2):
  # Step 1: Convert both rings to lowercase
  string1 = str1.lower()
  string2 = str2.lower()

  # Step 2: Sort the charactes of both strings
  sorted_string1 = sorted(string1)
  sorted_string2 = sorted(string2)

  # Step 3: Compare the sorted character lists
  if sorted_string1 == sorted_string2:
    return True
  else:
    return False
```

```python
# Example usage
result1 = is_anagram("listen", "silent")
print(result1)   # True
```

```python
result2 = is_anagram("hello", "world")
print(result2)   # False
```
