```python
# CATEGORICAL ENCODING
#gender:label encoding
#water_source: one-hot encoding
#Disease:Label encoding
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['Disease'] = le.fit_transform(df['Disease'])

df = pd.get_dummies(df, columns=['Water_Source'], drop_first=True)
```

```python
# FEATURE SCALING
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Separate features and target FIRST
X = df.drop('Disease', axis=1)
y = df['Disease']

# Select only numerical columns from X
numerical_cols = X.select_dtypes(include=['int64','float64']).columns

# Apply scaling only on numerical features
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])
```
