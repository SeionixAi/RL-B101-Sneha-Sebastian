# Pandas_class1

```python
import pandas as pd
```

1.Create DataFrame

```python
data = {"Name": ["Alice", "Bob", "Charlie"],"Age": [22, 24, 23],"Marks": [85, 90, 88]}
df = pd.DataFrame(data)
print(df)
```

3.Basic Operations

```python
print(df["Age"].mean())
print(df["Marks"].max())
```

4. Filtering

```python
print(df[df["Marks"] > 85])
```
